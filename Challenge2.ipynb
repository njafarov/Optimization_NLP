```python
!pip install pyomo
!wget -N -q "https://ampl.com/dl/open/ipopt/ipopt-linux64.zip"
!unzip -o -q ipopt-linux64
import pyomo.environ as pyo
from pyomo.opt import SolverFactory
```

```python
model = pyo.ConcreteModel()

#Decision Variables
model.X1 = pyo.Var(domain=pyo.NonNegativeReals)
X1 = model.X1
model.X2 = pyo.Var(domain=pyo.NonNegativeReals)
X2 = model.X2
model.X3 = pyo.Var(domain=pyo.NonNegativeReals)
X3 = model.X3
model.Y1 = pyo.Var(domain=pyo.NonNegativeReals)
Y1 = model.Y1
model.Y2 = pyo.Var(domain=pyo.NonNegativeReals)
Y2 = model.Y2
model.Y3 = pyo.Var(domain=pyo.NonNegativeReals)
Y3 = model.Y3
model.A = pyo.Var(domain=pyo.NonNegativeReals)
A = model.A
model.B = pyo.Var(domain=pyo.NonNegativeReals)
B = model.B

#Objective Function
model.Objf = pyo.Objective(expr = 2*(A+B),sense = pyo.minimize)

#Constraints
model.Const1 = pyo.Constraint(expr= X1>=6)
model.Const2 = pyo.Constraint(expr= Y1>=6)
model.Const3 = pyo.Constraint(expr= X2>=12)
model.Const4 = pyo.Constraint(expr= Y2>=12)
model.Const5 = pyo.Constraint(expr= X3>=16)
model.Const6 = pyo.Constraint(expr= Y3>=16)
model.Const7 = pyo.Constraint(expr= X1-B<=-6)
model.Const8 = pyo.Constraint(expr= Y1-A<=-6)
model.Const9 = pyo.Constraint(expr= X2-B<=-12)
model.Const10 = pyo.Constraint(expr= Y2-A<=-12)
model.Const11 = pyo.Constraint(expr= X3-B<=-16)
model.Const12 = pyo.Constraint(expr= Y3-A<=-16)
model.Const13 = pyo.Constraint(expr = (X1-X2)**2+(Y1-Y2)**2>=324)
model.Const14 = pyo.Constraint(expr = (X1-X3)**2+(Y1-Y3)**2>=484)
model.Const15 = pyo.Constraint(expr = (X2-X3)**2+(Y2-Y3)**2>=784)

Solver = SolverFactory('ipopt',executable = '/content/ipopt')
results = Solver.solve(model)

print(results)
print('Objective Function= ', model.Objf())
print('X1= ',X1())
print('Y1= ',Y1())
print('X2= ',X2())
print('Y2= ',Y2())
print('X3= ',X3())
print('Y3= ',Y3())
print('A= ',A())
print('B= ',B())
```
